# Distribución t de <i>Student</i>

Una variable aleatoria t de <i>t de Student</i>, con <i>m</i>, es la razón (ratio) de dos variables aleatorias, a saber: $\mathcal{Z} \sim \mathcal{N}(0, 1)$ y $\mathcal{W} \sim \chi_{m}^2$, que además son independientes entre si:

$$\frac{\mathcal{Z}}{{\sqrt{\frac{\mathcal{W}}{{m}}}}} \sim \mathcal{t_m}$$

La distribución es simétrica acampanada; puede verse más o menos achatada, dependiendo de los grados de libertad; y se aproxima a una normal cuando se tienen al menos 30 grados de libertad. 

```python
import numpy as np
from scipy.stats import *
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['text.usetex'] = True
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['text.usetex'] = True
```

```python
fig, ax = plt.subplots()

plt.title("Aproximación entre Distribuciones $t$\ny Normal Estándar")

gl = [2, 5, 25]
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99))

ax.plot(x, norm.pdf(x), lw=2, label='$\mathcal{N}(0, 1)$')

for i in gl:
    ax.plot(x, t.pdf(x, df=i), linestyle='--', alpha=0.6, label=f'$t-{i}$' + ' gl')
    plt.legend()
```

A medida que se incrementan los grados de libertad (gl), las trayectorias de las <i>t de Student</i>, se aproximan más a una normal estándar.
