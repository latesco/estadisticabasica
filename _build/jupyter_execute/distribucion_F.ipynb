# Distribución F

La <i>ratio</i> de dos variables aleatorias $\mathcal{X}$ y $\mathcal{Y}$, distribuidas ambas según una $\chi^2$ con grados de libertad <i>m</i> y <i>n</i> respectivamente, es lo que se conoce como una distribución $\mathcal{F_{m,n}}$:

$$\frac{\frac{X}{{m}}}{{\frac{Y}{{n}}}} \sim \mathcal{F_{m,n}}$$

$$con\;\;\mathcal{X} \sim \chi_m^2\;y\;\mathcal{Y} \sim \chi_n^2$$



Para generar números aleatorios, puede usarse <i>scipy</i> o <i>NumPy</i>

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

La probabilidad de que una variable $Y \sim \mathcal{F_{4,15}}$ tome valores superiores a 1.5, es decir:

$$P(Y \geq 1.5) = 1 - P(Y \lt 1.5)$$

Se obtiene con <i>scipy</i>:

```python
1 - stats.f.cdf(1.5, 4, 15)
```

Podemos generar valores aleatorios, para unos grados de libertad dados, y que sigan la distribución $\mathcal{F}$, y asi graficar su función de densidad.

```python
gl1=4
gl2=15
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['text.usetex'] = True
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['text.usetex'] = True
```

```python
ex=r"$P(Y \geq 1.5) \approx 0.25$"
```

```python
fig, ax = plt.subplots()

x = np.linspace(stats.f.ppf(0.01, dfn=gl1, dfd=gl2),
                stats.f.ppf(0.99, dfn=gl1, dfd=gl2), 100)

p = stats.f.pdf(x, dfn=gl1, dfd=gl2)

ax.plot(x, p,
       linestyle='-')
ax.fill_between(x, p, where = (x > 1.5),
               color = 'orange')
plt.margins(x=0.01)
plt.margins(y=0.01)
plt.title('Distribución $\mathcal{F}_{4,15}$')
plt.xlabel('x')
plt.ylabel('Probabilidad')
ax.annotate(ex, (2, 0.08), xytext=(120, 50),
           textcoords='offset points',
           fontsize=15,
           arrowprops=dict(
            arrowstyle='fancy',
            color='#000000',
            connectionstyle='arc3, rad=0.3'
            ))
```
